```python
import numpy as np
from scipy.optimize import bisect
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Constants
e = 3.795                # e^2 in eV·Å
hbar_c = 1973            # ℏc in eV·Å
m = 0.511e6              # Reduced mass of electron in eV/c^2

# Coulomb potential energy function
def V(r):
    return -e**2 / r  # Coulomb potential

# Pre-computation function A(r, E)
def A(r, E):
    return (2 * m / hbar_c**2) * (V(r) - E)

# Schrödinger equation model (reduced radial equation)
def schrodinger(w, r, E):
    u, v = w  # u = wave function, v = first derivative of u
    du_dr = v
    dv_dr = A(r, E) * u
    return [du_dr, dv_dr]

# Solve the Schrödinger equation for a given energy
def solve_wavefunction(E, r, w0):
    sol = odeint(schrodinger, w0, r, args=(E,))
    u = sol[:, 0]  # Extract the wave function u(r)
    return u

# Find eigenvalues using bisection method
def find_eigenvalues(energy_range, r, w0):
    eigenvalues = []  # List to store eigenvalues
    for i in range(len(energy_range) - 1):
        # Evaluate wavefunction at the boundary for two trial energies
        u1 = solve_wavefunction(energy_range[i], r, w0)[-1]
        u2 = solve_wavefunction(energy_range[i + 1], r, w0)[-1]
        # If the sign changes, there is an eigenvalue
        if np.sign(u1) != np.sign(u2):
            eigenvalue = bisect(lambda E: solve_wavefunction(E, r, w0)[-1],
                                energy_range[i], energy_range[i + 1])
            eigenvalues.append(eigenvalue)
        if len(eigenvalues) == 2:  # Stop after 2 states
                break
    return eigenvalues

# Plot wave functions for the found eigenvalues
def plot_wavefunctions(eigenvalues,r,w0):
    plt.figure(figsize=(10, 6))
    for i, E in enumerate(eigenvalues):
        u = solve_wavefunction(E, r, w0)
        plt.plot(r, u, label=f"State {i+1}: E = {E:.2f} eV")
    plt.title("Radial Wave Functions for Hydrogen Atom for ground state and first state")
    plt.xlabel("Radial Distance r (Å)")
    plt.ylabel("Wave Function u(r)")
    plt.legend()
    plt.grid()
    plt.show()

# Main Code
if __name__ == "__main__":
    # Radial range (avoid r=0 singularity, integrate to large r)
    r = np.linspace(1e-3, 10, 1000)  # r from 0.001 to 20 Å

    # Initial conditions: small u(0) and du/dr(0)
    w0 = [1e-5, 1]

    # Energy search range (approximate hydrogen energy levels)
    energy_range = np.linspace(-20, 0, 200)  # Energies from -20 eV to 0 eV

    # Find eigenvalues
    for i, E in enumerate(eigenvalues): # Use enumerate to get index i
        eigenvalues = find_eigenvalues(En, r, w0)
        print(f"State {i+1}: E = {E:.2f} eV") # Corrected indentation
    # Plot wave functions for the found eigenvalues
    plot_wavefunctions(eigenvalues, r, w0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Physical constants
h = 6.626e-34  # Planck's constant (J·s)
c = 3.0e8      # Speed of light (m/s)
k_B = 1.381e-23  # Boltzmann constant (J/K)

# Frequency range (in Hz), from 10^12 to 10^15 Hz
nu = np.logspace(12, 15, 500)

# Temperatures to compare (in Kelvin)
T_low = 300    # Low temperature (e.g., room temp)
T_high = 6000  # High temperature (e.g., Sun-like)

# Planck's Law: Spectral radiance B_nu(T) in W/(m^2·Hz·sr)
def planck_law(nu, T):
    numerator = 2 * h * nu**3 / c**2
    denominator = np.exp(h * nu / (k_B * T)) - 1
    return numerator / denominator

# Rayleigh-Jeans Law: Classical approximation
def rayleigh_jeans_law(nu, T):
    return 2 * nu**2 * k_B * T / c**2

# Calculate for low temperature
B_planck_low = planck_law(nu, T_low)
B_rj_low = rayleigh_jeans_law(nu, T_low)

# Calculate for high temperature
B_planck_high = planck_law(nu, T_high)
B_rj_high = rayleigh_jeans_law(nu, T_high)

# Plotting
plt.figure(figsize=(12, 8))

# Low temperature plot
plt.subplot(2, 1, 1)
plt.loglog(nu, B_planck_low, label="Planck's Law", color="blue")
plt.loglog(nu, B_rj_low, label="Rayleigh-Jeans Law", color="red", linestyle="--")
plt.xlabel("Frequency ν (Hz)")
plt.ylabel("Spectral Radiance $B_ν$ (W/m²·Hz·sr)")
plt.title(f"Blackbody Radiation at T = {T_low} K (Low Temperature)")
plt.legend()
plt.grid(True, which="both", ls="--")
plt.ylim(1e-20, 1e-10)  # Adjust y-limits for visibility

# High temperature plot
plt.subplot(2, 1, 2)
plt.loglog(nu, B_planck_high, label="Planck's Law", color="blue")
plt.loglog(nu, B_rj_high, label="Rayleigh-Jeans Law", color="red", linestyle="--")
plt.xlabel("Frequency ν (Hz)")
plt.ylabel("Spectral Radiance $B_ν$ (W/m²·Hz·sr)")
plt.title(f"Blackbody Radiation at T = {T_high} K (High Temperature)")
plt.legend()
plt.grid(True, which="both", ls="--")
plt.ylim(1e-10, 1e2)  # Adjust y-limits for visibility

# Adjust layout and show plot
plt.tight_layout()
plt.show()

# Print comparisons at specific frequencies
nu_sample = [1e13, 5e14]  # Example frequencies
print(f"\nAt T = {T_low} K (Low Temperature):")
for n in nu_sample:
    print(f"ν = {n:.1e} Hz:")
    print(f"  Planck: {planck_law(n, T_low):.2e} W/(m²·Hz·sr)")
    print(f"  R-J:    {rayleigh_jeans_law(n, T_low):.2e} W/(m²·Hz·sr)")

print(f"\nAt T = {T_high} K (High Temperature):")
for n in nu_sample:
    print(f"ν = {n:.1e} Hz:")
    print(f"  Planck: {planck_law(n, T_high):.2e} W/(m²·Hz·sr)")
    print(f"  R-J:    {rayleigh_jeans_law(n, T_high):.2e} W/(m²·Hz·sr)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const

h
def plancks_law(wavelength, temperature):
    """
    Calculate spectral radiance using Planck's law.
    :param wavelength: Wavelength in meters
    :param temperature: Temperature in Kelvin
    :return: Spectral radiance in W/m^3/sr
    """
    return (8 * np.pi * const.k * temperature) / (wavelength**4)

def rayleigh_jeans_law(wavelength, temperature):
    """
    Calculate spectral radiance using Rayleigh-Jeans law.
    """
    return (8 * np.pi * const.k * temperature) / (wavelength**4)

def wiens_law(wavelength, temperature):
    """
    Calculate spectral radiance using Wien's approximation.
    """
    return ((8 * np.pi * const.h * const.c) / (wavelength**5)) * np.exp(- (const.h * const.c) / (wavelength * const.k * temperature))

wavelengths = np.linspace(1e-7, 3e-6, 500)  # From 100 nm to 3000 nm

temperatures = [3000, 4000, 5000, 6000]  # Kelvin

plt.figure(figsize=(8, 6))
for T in temperatures:
    plt.plot(wavelengths * 1e9, plancks_law(wavelengths, T), label=f"Planck's Law (T = {T} K)")
    plt.plot(wavelengths * 1e9, rayleigh_jeans_law(wavelengths, T), linestyle='dashed', label=f"Rayleigh-Jeans Law (T = {T} K)")
    plt.plot(wavelengths * 1e9, wiens_law(wavelengths, T), linestyle='dotted', label=f"Wien's Law (T = {T} K)")

plt.xlabel("Wavelength (nm)")
plt.ylabel("Spectral Radiance (W/m^3/sr)")
plt.title("Black Body Radiation - Comparison of Laws")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from scipy import integrate

# Physical constants
k_B = 1.380649e-23  # Boltzmann constant (J/K)
N_A = 6.02214076e23  # Avogadro's number
R = k_B * N_A  # Gas constant (J/mol·K) = 8.31446 J/mol·K

# Characteristic temperatures (example values)
theta_E = 200  # Einstein temperature (K)
theta_D = 250  # Debye temperature (K)

# Dulong-Petit Law - constant 3R
def dulong_petit():
    return 3 * R

# Einstein model
def einstein_model(T):
    x = theta_E / T
    exp_x = np.exp(x)
    return 3 * R * (x**2) * exp_x / ((exp_x - 1)**2)

# Debye integral term
def debye_integrand(x):
    return (x**4) * np.exp(x) / ((np.exp(x) - 1)**2)

# Debye model
def debye_model(T):
    x_max = theta_D / T
    if T == 0:
        return 0  # Avoiding division by zero

    # For very low temperatures, use the T^3 law approximation
    if x_max > 100:  # Arbitrary large value to switch to approximation
        return (12 * np.pi**4 * R) / (5 * (theta_D**3)) * (T**3)

    # Numerical integration for regular temperatures
    integral, _ = integrate.quad(debye_integrand, 0, x_max)
    return 9 * R * integral / (x_max**3)

# Temperature ranges (K)
T_low = np.linspace(1, 100, 200)  # Low temperature range (avoid T=0 for numerical stability)
T_high = np.linspace(100, 1000, 200)  # High temperature range
T_full = np.linspace(1, 1000, 400)  # Full temperature range for combined plot

# Calculate specific heat for each model
# Low temperature
c_dp_low = np.full_like(T_low, dulong_petit())
c_einstein_low = np.array([einstein_model(t) for t in T_low])
c_debye_low = np.array([debye_model(t) for t in T_low])

# High temperature
c_dp_high = np.full_like(T_high, dulong_petit())
c_einstein_high = np.array([einstein_model(t) for t in T_high])
c_debye_high = np.array([debye_model(t) for t in T_high])

# Full range
c_dp_full = np.full_like(T_full, dulong_petit())
c_einstein_full = np.array([einstein_model(t) for t in T_full])
c_debye_full = np.array([debye_model(t) for t in T_full])

# Create figure with subplots
fig = plt.figure(figsize=(12, 10))
gs = GridSpec(2, 2, figure=fig)
ax1 = fig.add_subplot(gs[0, :])  # Top row, spans both columns
ax2 = fig.add_subplot(gs[1, 0])  # Bottom left
ax3 = fig.add_subplot(gs[1, 1])  # Bottom right

# Full temperature range plot
ax1.plot(T_full, c_dp_full, 'k-', label="Dulong-Petit Law")
ax1.plot(T_full, c_einstein_full, 'r-', label="Einstein Model")
ax1.plot(T_full, c_debye_full, 'b-', label="Debye Model")

# Add vertical lines marking characteristic temperatures
ax1.axvline(x=theta_E, color='r', linestyle='--', alpha=0.5, label=f"Einstein Temperature ({theta_E} K)")
ax1.axvline(x=theta_D, color='b', linestyle='--', alpha=0.5, label=f"Debye Temperature ({theta_D} K)")

ax1.set_title("Specific Heat of Solids: Comparison of Models")
ax1.set_xlabel("Temperature (K)")
ax1.set_ylabel("Specific Heat (J/mol·K)")
ax1.legend()
ax1.grid(True, alpha=0.3)

# Low temperature plot with log-log scale to show T³ behavior
ax2.loglog(T_low, c_dp_low, 'k-', label="Dulong-Petit Law")
ax2.loglog(T_low, c_einstein_low, 'r-', label="Einstein Model")
ax2.loglog(T_low, c_debye_low, 'b-', label="Debye Model")

# Add T³ reference line
T3_ref = T_low**3 * (c_debye_low[20] / (T_low[20]**3))
ax2.loglog(T_low, T3_ref, 'g--', label="T³ Law")

ax2.set_title("Low Temperature Behavior (Log-Log Scale)")
ax2.set_xlabel("Temperature (K)")
ax2.set_ylabel("Specific Heat (J/mol·K)")
ax2.legend()
ax2.grid(True, alpha=0.3, which="both")

# High temperature plot
ax3.plot(T_high, c_dp_high, 'k-', label="Dulong-Petit Law")
ax3.plot(T_high, c_einstein_high, 'r-', label="Einstein Model")
ax3.plot(T_high, c_debye_high, 'b-', label="Debye Model")

ax3.set_title("High Temperature Behavior")
ax3.set_xlabel("Temperature (K)")
ax3.set_ylabel("Specific Heat (J/mol·K)")
ax3.set_ylim(0, 30)  # Limit y-axis to better show convergence
ax3.legend()
ax3.grid(True, alpha=0.3)

# Add Dulong-Petit limit line
ax3.axhline(y=3*R, color='k', linestyle='--', alpha=0.5, label=f"3R = {3*R:.2f} J/mol·K")

plt.tight_layout()
plt.show()

# Create a plot showing normalized values to better visualize temperature dependence
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Normalize all models by 3R (Dulong-Petit value)
dp_norm = 1.0  # By definition

# Low temperature normalized
einstein_norm_low = c_einstein_low / (3*R)
debye_norm_low = c_debye_low / (3*R)

# High temperature normalized
einstein_norm_high = c_einstein_high / (3*R)
debye_norm_high = c_debye_high / (3*R)

# Low temperature plot (normalized)
ax1.semilogx(T_low, np.full_like(T_low, dp_norm), 'k-', label="Dulong-Petit Law")
ax1.semilogx(T_low, einstein_norm_low, 'r-', label="Einstein Model")
ax1.semilogx(T_low, debye_norm_low, 'b-', label="Debye Model")
ax1.set_title("Low Temperature (Normalized by 3R)")
ax1.set_xlabel("Temperature (K)")
ax1.set_ylabel("C_V / 3R")
ax1.legend()
ax1.grid(True, alpha=0.3)

# High temperature plot (normalized)
ax2.plot(T_high, np.full_like(T_high, dp_norm), 'k-', label="Dulong-Petit Law")
ax2.plot(T_high, einstein_norm_high, 'r-', label="Einstein Model")
ax2.plot(T_high, debye_norm_high, 'b-', label="Debye Model")
ax2.set_title("High Temperature (Normalized by 3R)")
ax2.set_xlabel("Temperature (K)")
ax2.set_ylabel("C_V / 3R")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k_B = 1.380649e-23  # Boltzmann constant (J/K), but we'll use k_B = 1 for simplicity
epsilon = 1.0       # Energy level spacing (arbitrary units)
N = 2               # Number of particles (adjustable)

# Temperature range (in units of epsilon / k_B)
T = np.linspace(0.01, 10, 200)  # k_B T / epsilon
beta = 1 / T            # beta * epsilon

# Maxwell-Boltzmann Statistics
def mb_partition(N, beta):
    z = 1 + np.exp(-beta)  # Single-particle partition function
    return z**N

def mb_energy(N, beta):
    return N * epsilon * np.exp(-beta) / (1 + np.exp(-beta))

def mb_fluctuation(N, beta):
    term = epsilon**2 * N * np.exp(-beta) / (1 + np.exp(-beta))**2
    return np.sqrt(term)

def mb_specific_heat(N, beta):
    x = beta
    return N * (epsilon)**2 * np.exp(-x) / (k_B * T**2 * (1 + np.exp(-x))**2)

# Fermi-Dirac Statistics (N <= 2 for 2 levels)
def fd_partition(N, beta):
    if N == 0:
        return 1
    elif N == 1:
        return 1 + np.exp(-beta)
    elif N == 2:
        return 1  # Only one state: (1, 1)
    else:
        raise ValueError("N > 2 not possible for FD with 2 levels")

def fd_energy(N, beta):
    if N == 1:
        return epsilon * np.exp(-beta) / (1 + np.exp(-beta))
    elif N == 2:
        return epsilon  # Fixed energy
    return 0

def fd_fluctuation(N, beta):
    if N == 1:
        return epsilon * np.sqrt(np.exp(-beta) / (1 + np.exp(-beta))**2)
    return 0  # N = 2 has no fluctuation (fixed state)

def fd_specific_heat(N, beta):
    if N == 1:
        x = beta
        return (epsilon)**2 * np.exp(-x) / (k_B * T**2 * (1 + np.exp(-x))**2)
    return 0

# Bose-Einstein Statistics
def be_partition(N, beta):
    return (1 - np.exp(-beta * (N + 1))) / (1 - np.exp(-beta))

def be_energy(N, beta):
    return epsilon * (np.exp(-beta) / (1 - np.exp(-beta)) -
                     (N + 1) * np.exp(-beta * (N + 1)) / (1 - np.exp(-beta * (N + 1))))

def be_fluctuation(N, beta):
    term1 = epsilon**2 * np.exp(-beta) / (1 - np.exp(-beta))**2
    term2 = epsilon**2 * (N + 1)**2 * np.exp(-beta * (N + 1)) / (1 - np.exp(-beta * (N + 1)))**2
    return np.sqrt(term1 + term2)

def be_specific_heat(N, beta):
    x = beta
    term1 = (epsilon)**2 * np.exp(-x) / (k_B * T**2 * (1 - np.exp(-x))**2)
    term2 = (N + 1)**2 * (epsilon)**2 * np.exp(-x * (N + 1)) / (k_B * T**2 * (1 - np.exp(-x * (N + 1)))**2)
    return term1 - term2

# Occupation numbers
def mb_n1(N, beta):
    return N * np.exp(-beta) / (1 + np.exp(-beta))

def fd_n1(N, beta):
    if N == 1:
        return 1 / (1 + np.exp(beta))
    elif N == 2:
        return 1
    return 0

def be_n1(N, beta):
    return 1 / (np.exp(beta) - 1)

# Compute quantities
Z_mb = mb_partition(N, beta)
E_mb = mb_energy(N, beta)
dE_mb = mb_fluctuation(N, beta)
Cv_mb = mb_specific_heat(N, beta)
n1_mb = mb_n1(N, beta)
n0_mb = N - n1_mb

# Calculate FD quantities for each temperature
Z_fd = np.array([fd_partition(N, b) for b in beta])  # Vectorized calculation
E_fd = np.array([fd_energy(N, b) for b in beta])       # Vectorized calculation
dE_fd = np.array([fd_fluctuation(N, b) for b in beta])   # Vectorized calculation
Cv_fd = np.array([fd_specific_heat(N, b) for b in beta])  # Vectorized calculation
n1_fd = np.array([fd_n1(N, b) for b in beta])        # Vectorized calculation
n0_fd = N - n1_fd

Z_be = be_partition(N, beta)
E_be = be_energy(N, beta)
dE_be = be_fluctuation(N, beta)
Cv_be = be_specific_heat(N, beta)
n1_be = be_n1(N, beta)
n0_be = N - n1_be

# Plotting
plt.figure(figsize=(15, 10))

# Average Energy
plt.subplot(2, 2, 1)
plt.plot(T, E_mb, label='MB')
plt.plot(T, E_fd, label='FD')
plt.plot(T, E_be, label='BE')
plt.xlabel(r'$k_B T / \epsilon$')
plt.ylabel(r'$\langle E \rangle / \epsilon$')
plt.title('Average Energy')
plt.legend()
plt.grid(True)

# Energy Fluctuation
plt.subplot(2, 2, 2)
plt.plot(T, dE_mb, label='MB')
plt.plot(T, dE_fd, label='FD')
plt.plot(T, dE_be, label='BE')
plt.xlabel(r'$k_B T / \epsilon$')
plt.ylabel(r'$\Delta E / \epsilon$')
plt.title('Energy Fluctuation')
plt.legend()
plt.grid(True)

# Specific Heat
plt.subplot(2, 2, 3)
plt.plot(T, Cv_mb, label='MB')
plt.plot(T, Cv_fd, label='FD')
plt.plot(T, Cv_be, label='BE')
plt.xlabel(r'$k_B T / \epsilon$')
plt.ylabel(r'$C_V / k_B$')
plt.title('Specific Heat')
plt.legend()
plt.grid(True)

# Occupation Number Ratio (n1 / n0)
plt.subplot(2, 2, 4)
plt.plot(T, n1_mb / n0_mb, label='MB')
plt.plot(T, n1_fd / n0_fd, label='FD')
plt.plot(T, n1_be / n0_be, label='BE')
plt.xlabel(r'$k_B T / \epsilon$')
plt.ylabel(r'$\langle n_1 \rangle / \langle n_0 \rangle$')
plt.title('Occupation Number Ratio')
plt.legend()
plt.grid(True)
plt.yscale('log')  # Log scale to see low T behavior

plt.tight_layout()
plt.show()

# Print values at low and high T
T_low, T_high = 0.1, 10.0
beta_low, beta_high = 1 / T_low, 1 / T_high
print(f"\nAt T = {T_low} (k_B T / \epsilon):")
print(f"MB: E = {mb_energy(N, beta_low):.3f}, Cv = {mb_specific_heat(N, beta_low):.3f}, n1/n0 = {mb_n1(N, beta_low)/ (N - mb_n1(N, beta_low)):.3e}")
print(f"FD: E = {fd_energy(N, beta_low):.3f}, Cv = {fd_specific_heat(N, beta_low):.3f}, n1/n0 = {fd_n1(N, beta_low)/ (N - fd_n1(N, beta_low)):.3e}")
print(f"BE: E = {be_energy(N, beta_low):.3f}, Cv = {be_specific_heat(N, beta_low):.3f}, n1/n0 = {be_n1(N, beta_low)/ (N - be_n1(N, beta_low)):.3e}")

print(f"\nAt T = {T_high} (k_B T / \epsilon):")
print(f"MB: E = {mb_energy(N, beta_high):.3f}, Cv = {mb_specific_heat(N, beta_high):.3f}, n1/n0 = {mb_n1(N, beta_high)/ (N - mb_n1(N, beta_high)):.3f}")
print(f"FD: E = {fd_energy(N, beta_high):.3f}, Cv = {fd_specific_heat(N, beta_high):.3f}, n1/n0 = {fd_n1(N, beta_high)/ (N - fd_n1(N, beta_high)):.3f}")
print(f"BE: E = {be_energy(N, beta_high):.3f}, Cv = {be_specific_heat(N, beta_high):.3f}, n1/n0 = {be_n1(N, beta_high)/ (N - be_n1(N, beta_high)):.3f}")
```
